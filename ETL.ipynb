```python
import pandas as pd
from geopy import distance
newport_ri = (41.49008, -71.312796)
cleveland_oh = (41.499498, -81.695391)
print(distance.distance(newport_ri, cleveland_oh).miles)
```

```python
month_list = ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10", "11", "12"]
df = {}
for month in month_list:
    file_name = f"Resources/JC-2020{month}-citibike-tripdata.csv"
    # print(file_name)
    df[month] = pd.read_csv(file_name)
```

```python
df_combined = df["01"]
for i in range(1, 12):
    month = month_list[i]
    df_combined = pd.concat([df_combined, df[month]])
print("shape: " + str(df_combined.shape))
```

```python
def calculate_distance(row):
    start = (row["start station latitude"], row["start station longitude"])
    end = (row["end station latitude"],row["end station longitude"])
    return distance.distance(start, end).miles
```

```python
df_combined["distance (miles)"] = df_combined.apply(calculate_distance, axis=1)
print("shape: " + str(df_combined.shape))
df_combined.head()
```

```python
df_combined.to_csv("Resources/JC-2020-citibike-tripdata-combined.csv")
```
